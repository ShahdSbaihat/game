```python

import random


adj_tuple = ("Brave", "Gentle", "Loud", "Soft", "Shiny", "Bitter", "Friendly", "Old", "Colorful", "Sleepy")
adv_tuple = ("Quickly", "Slowly", "Carefully", "Happily", "Quietly", "Loudly", "Easily", "Well", "Badly", "Clearly")
noun_tuple = ("Time", "Music", "Phone", "Water", "House", "Cat", "Ball", "Chair", "Flower", "Computer")
verb_tuple = ("Run", "Eat", "Jump", "Write", "Speak", "Dance", "Sing", "Read", "Sleep", "Swim")

tuples_list = adj_tuple + adv_tuple + noun_tuple + verb_tuple
words_list=[]

def select_random_words(tuples_list):
    for i in range(10):
        random_word = random.choice(tuples_list)
        words_list.append(random_word)



def calculate_points(word):
    a = 0
    if word in noun_tuple:
        a = 5
    elif word in verb_tuple:
        a = 7
    elif word in adj_tuple:
        a = 2
    elif word in adv_tuple:
        a = 1
    return a


def display_score(player_list, computer_list,player_last):
    player_score=0
    computer_score=0
    if (player_last==True):
        player_score += 5
    else:
        computer_score += 5
    for word in player_list:
        player_score += (calculate_points(word))
    for word in computer_list:
        computer_score += (calculate_points(word))
    print("Player score:", player_score)
    print("Computer score:", computer_score)


def play_game():
    player_list = []
    computer_list = []
    select_random_words(tuples_list)
    rnum1 = random.randint(1, 2)
    attemp_count = 0
    player_last=False
    while (True and rnum1==1):

        print("Player's turn:")
        attemp_count += 1
        if (attemp_count==10):
            player_last=True
        print("Your list:", player_list)
        index = int(input("Choose the index of the word you want to add to your list (1-"+str(len(words_list))+"), or -1 to return a word: "))
        if index == -1:
            print("Returning a word from your list.")
            attemp_count -= 1
            print("Your list:", player_list)
            return_index = int(input("Choose the index of the word you want to return (-1 to cancel): "))
            if return_index != -1:

                if return_index >= 1 and return_index < len(player_list):
                    returned_word = player_list[return_index]
                    player_list.pop(return_index - 1)
                    words_list.append(returned_word)
                    print("Returned word:", returned_word)

                else:
                    print("Invalid index.")
                    continue
            else:
                continue
        elif index >= 1 and index <= 10:
            chosen_word = words_list.pop(index - 1)
            player_list.append(chosen_word)
            print("Chosen word:", chosen_word)
            print("Your list:", player_list)
        else:
            print("Invalid index.")
            continue

        if all(word == "" for word in words_list):
            print("Words list is empty. Game over!")
            break

        print("\nComputer's turn:")
        attemp_count += 1
        index = random.randint(1, len(words_list))
        chosen_word = words_list.pop(index - 1)
        computer_list.append(chosen_word)
        print("Chosen word by computer:", chosen_word)
        print("Computer's list:", computer_list)

        if all(word == "" for word in words_list):
            print("Words list is empty. Game over!")
            break

    while (True and rnum1==2):

        print("\nComputer's turn:")
        attemp_count += 1
        index = random.randint(1, len(words_list))
        chosen_word = words_list.pop(index - 1)
        computer_list.append(chosen_word)
        print("Chosen word by computer:", chosen_word)
        print("Computer's list:", computer_list)

        if all(word == "" for word in words_list):
            print("Words list is empty. Game over!")
            break

        print("Player's turn:")
        attemp_count += 1
        if (attemp_count==10):
            player_last=True
        print("Your list:", player_list)
        index = int(input("Choose the index of the word you want to add to your list (1-"+str(len(words_list))+"), or -1 to return a word: "))
        if index == -1:
            print("Returning a word from your list.")
            attemp_count -= 1
            print("Your list:", player_list)
            return_index = int(input("Choose the index of the word you want to return (-1 to cancel): "))
            if return_index != -1:
                if return_index >= 1 and return_index < len(player_list):
                    returned_word =player_list[return_index]
                    player_list.pop(return_index - 1)
                    words_list.append(returned_word)
                    print("Returned word:", returned_word)
                else:
                    print("Invalid index.")
                    continue
            else:
                continue
        elif index >= 1 and index <= 10:
            chosen_word = words_list.pop(index - 1)
            player_list.append(chosen_word)
            print("Chosen word:", chosen_word)
            print("Your list:", player_list)
        else:
            print("Invalid index.")
            continue

        if all(word == "" for word in words_list):
            print("Words list is empty. Game over!")
            break

    print("\nFinal scores:")
    display_score(player_list, computer_list,player_last)




play_game()
```
